```python
#Creating a Student Registration form GUI using tkinter

from tkinter import *
from tkinter import messagebox
import re
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

root = Tk()
root.title("Systecks Solutions")
root.geometry("460x500")
root.resizable(False,False)

#creating funtions

def register():
    name=name_box.get().strip()
    aicte=aicte_box.get().strip()
    email=email_box.get().strip()
    phone=phone_box.get().strip()
    college=college_box.get().strip()

 # Validate inputs
    if not name:
        messagebox.showerror("Error", "Please enter your name.")
        return
    if not aicte:
        messagebox.showerror("Error", "Please enter an AICTE ID.")
        return
    if not re.match(r"[^@]+@[^@]+\.[^@]+", email) or not email.endswith("@gmail.com"):
        messagebox.showerror("Error", "Please enter a valid Gmail/email address.")
        return
    if not phone.isdigit() or len(phone) != 10:
        messagebox.showerror("Error", "Please enter a valid 10-digit phone number.")
        return
    if not college:
        messagebox.showerror("Error", "Please enter your college name.")
        return

    generate_pdf(name, aicte, email, phone, college)
    messagebox.showinfo("Success", "Successfully Registered")


def generate_pdf(name, aicte, email, phone, college_name):
    # Create PDF document
    pdf_filename = "Student_data.pdf"
    details = canvas.Canvas(pdf_filename, pagesize=letter)
    
    # Write data to PDF
    details.drawString(100, 680, "STUDENT DETAILS:")
    details.drawString(100, 650, f"Name: {name}")
    details.drawString(100, 630, f"AICTE ID: {aicte}")
    details.drawString(100, 610, f"Email ID: {email}")
    details.drawString(100, 590, f"Phone No: {phone}")
    details.drawString(100, 570, f"College Name: {college_name}")
    
    details.save()


#Creating Student's details needed for Registration

Label(root, text="Student Registration form", font="ariel 20 bold", bg="red",fg="white").pack(fill="both")
Label(root, text="Name", font="20").place(x=40,y=90)
Label(root, text="AICTE ID", font="20").place(x=40,y=130)
Label(root, text="Email ID", font="20").place(x=40,y=170)
Label(root, text="Phone No", font="20").place(x=40,y=210)
Label(root, text="College Name", font="20").place(x=40,y=250)


#creating inputs for the Labels

name_box = Entry(root, font="10", bd=4)
name_box.place(x=180, y=90)
aicte_box = Entry(root, font="10", bd=4)
aicte_box.place(x=180, y=130)
email_box = Entry(root, font="10", bd=4)
email_box.place(x=180, y=170)
phone_box = Entry(root, font="10", bd=4)
phone_box.place(x=180, y=210)
college_box = Entry(root, font="10", bd=4)
college_box.place(x=180, y=250)

Button(text = "Register",font="20", command= register).place(x=240, y=310)

root.mainloop()
```
